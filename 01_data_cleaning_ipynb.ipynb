```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("adarshsng/lending-club-loan-data-csv")

print("Path to dataset files:", path)
```

```python
keep_cols = [
    "loan_status",
    "loan_amnt",
    "term",
    "int_rate",
    "installment",
    "grade",
    "sub_grade",
    "annual_inc",
    "emp_length",
    "home_ownership",
    "verification_status",
    "purpose",
    "dti",
    "earliest_cr_line",
    "open_acc",
    "total_acc",
    "inq_last_6mths",
    "pub_rec",
    "revol_bal",
    "revol_util",
    "addr_state",
    "issue_d",
    "application_type"
]
```

```python
df = pd.read_csv(
    "/kaggle/input/lending-club-loan-data-csv/loan.csv",
    usecols=keep_cols
)
```

```python
df[keep_cols].to_csv("credit_risk_selected_vars.csv", index=False)
```
