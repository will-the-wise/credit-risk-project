```python
df['annual_inc'] = df['annual_inc'].fillna(df['annual_inc'].median())
```

```python
df['emp_length'] = df['emp_length'].replace({
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
})
```

```python
df['emp_length'] = df['emp_length'].fillna(df['emp_length'].median())
```

```python
df['revol_util'] = df['revol_util'].fillna(df['revol_util'].median())
df['revol_util'] = df['revol_util'].clip(upper=100)
```

```python
df['dti'] = df['dti'].fillna(df['dti'].median())
```

```python
cat_vars = [
    'home_ownership','verification_status',
    'purpose','addr_state','application_type'
]

for col in cat_vars:
    df[col] = df[col].fillna('Unknown')
```

```python
def winsorize(series, lower=0.01, upper=0.99):
    return series.clip(
        lower=series.quantile(lower),
        upper=series.quantile(upper)
    )
```

```python
for col in ['annual_inc','loan_amnt','revol_bal']:
    df[col] = winsorize(df[col])
```

```python
grade_map = {
    'A': 1, 'B': 2, 'C': 3,
    'D': 4, 'E': 5, 'F': 6, 'G': 7
}

df['grade_num'] = df['grade'].map(grade_map)
```

```python
df = df.drop(columns=['sub_grade'])
```

```python
cat_vars = [
    'purpose',
    'home_ownership',
    'verification_status',
    'addr_state',
    'application_type'
]

df = pd.get_dummies(df, columns=cat_vars, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scale_vars = [
    'annual_inc','loan_amnt','installment',
    'int_rate','dti','revol_bal','revol_util'
]

scaler = StandardScaler()
df[scale_vars] = scaler.fit_transform(df[scale_vars])
```

```python
df = df.drop(columns=['grade'])
```
